```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
model=ChatGroq(model="llama-3.1-8b-instant",api_key=groq_api_key)
model
```

```python
parser=StrOutputParser()
```

```python
from langchain_core.prompts import ChatPromptTemplate

generic_template="You are a language translation friendly chat AI. Translate {input} into {language}. Just give the translated text as the only response and no fillers."

prompt=ChatPromptTemplate.from_messages(
    [
        ("system",generic_template),
        ("user","{input}")
    ]
)
```

```python
chain=prompt|model|parser
chain.invoke({"language":"Hindi","input":"Hello, How are you?"})
```
